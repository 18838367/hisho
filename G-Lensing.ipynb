```python
def pointSource(theta):
    tgeom=(1+rshft)*Dd*Ds/(Dds*C)*(0.5*(theta-beta)**2)
    tgrav=(1+rshft)*Dd*Ds/(Dds*C)*(2*Dds*G*M*np.abs(np.arcsinh(Dds/(theta*Dd))))/(Ds*Dd*C**2)
    t=tgrav+tgeom
    return t, tgrav, tgeom
```
