```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math
```

```python
t=np.arange(100)/100*(2*math.pi)
```

```python
#A sine and cosine with frequencies that are multiples of the same fundamental frequency, multiplied together 
#will always be symmetric about zero and will therefore always have a zero period integral
plt.plot(t, np.sin(t), t, np.cos(2*t), t, np.cos(2*t)*np.sin(t))
plt.show()
```

```python
#A cosine (or sine) and a cosine (or sine) with frequencies that are multiples of the same fundamental frequency,
#multiplied together are always symmetric about zero and will therefore always have a zero period integral except 
#if their frequencies are the same
plt.plot(t, np.sin(t), t, np.sin(2*t+2), t, np.sin(2*t+2)*np.sin(t))
plt.show()
plt.plot(t, np.sin(t), t, np.sin(t), t, np.sin(t)*np.sin(t))
plt.show()
plt.plot(t, np.cos(t), t, np.cos(2*t+2), t, np.cos(2*t+2)*np.cos(t))
plt.show()
plt.plot(t, np.cos(t), t, np.cos(t), t, np.cos(t)*np.cos(t))
plt.show()
```

```python
######diffractive scintillation test
def pathLength(L, theta0):
    D=(L**2*theta0**2+L**2)**(0.5)-L
    plt.plot(theta0,D)
```

```python
theta0=np.arange(0,1000)/1000*math.pi/2
pathLength(10000,theta0)
```
